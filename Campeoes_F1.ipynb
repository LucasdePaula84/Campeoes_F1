## Dados retirados do site wikipedia
https://pt.wikipedia.org/wiki/Lista_de_campe%C3%B5es_da_F%C3%B3rmula_1

```python
# Importar biblioteca
import pandas as pd
```

```python
# Extrair dados
tabelas = pd.read_html("https://pt.wikipedia.org/wiki/Lista_de_campe%C3%B5es_da_F%C3%B3rmula_1")
```

```python
# Selecionando a tabela desejada
tabela = tabelas[2]
tabela
```

```python
# Quantidade de titulos por piloto
campeoes = tabela['Piloto'].value_counts()
campeoes
```

```python
# Quantidade de titulos por Equipe
equipe = tabela["Equipe"].value_counts()
equipe
```

```python
# Quantidade de titulos por Motor
motor = tabela["Motor"].value_counts()
motor
```

```python
# Quantidade de Poles por piloto
poles_piloto = tabela.groupby("Piloto")["Poles"].sum().sort_values(ascending=False)
poles_piloto
```

```python
# Quantidade de Vitorias por piloto
vitorias_piloto = tabela.groupby("Piloto")["Vitórias"].sum().sort_values(ascending=False)
vitorias_piloto
```

```python
# Aproveitamento 
aproveitamento = tabela.groupby("Piloto")[["Poles", "Vitórias"]].sum()
aproveitamento["Aproveitamento"] = aproveitamento["Vitórias"] / aproveitamento["Poles"]
aproveitamento = aproveitamento.sort_values(by="Aproveitamento", ascending=False)
aproveitamento
```

```python
# Quantidade de titulos por marca de pneu
pneu = tabela['Pneu'].value_counts()
pneu
```
